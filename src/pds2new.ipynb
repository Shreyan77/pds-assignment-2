```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Load the dataset
df = pd.read_csv('/content/train.csv')

# Create a SimpleImputer for both categorical and numerical columns
imputer = SimpleImputer(strategy='most_frequent')  # For categorical columns
df[df.select_dtypes(include=['object']).columns] = imputer.fit_transform(df.select_dtypes(include=['object']))

imputer = SimpleImputer(strategy='mean')  # For numerical columns
df[df.select_dtypes(include=['number']).columns] = imputer.fit_transform(df.select_dtypes(include=['number']))

# Save the cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)
```

b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from Mileage, CC from Engine, bhp from Power, and lakh from
New_price).

```python
clean_data = pd.read_csv('/content/cleaned_dataset.csv')
```

```python
import pandas as pd

# Load the dataset
clean_data = pd.read_csv('/content/cleaned_dataset.csv')

# Define a dictionary to map units to empty strings for each column
unit_mapping = {
    'Mileage': ['kmpl', 'km/kg'],
    'Engine': ['CC'],
    'Power': ['bhp'],
    'New_Price': ['Lakh', 'Cr']
}

# Iterate through columns and remove units using a loop
for column, units in unit_mapping.items():
    for unit in units:
        clean_data[column] = clean_data[column].str.replace(unit, '')

# Convert the cleaned columns to float
columns_to_clean = ['Mileage', 'Engine', 'Power', 'New_Price']
clean_data[columns_to_clean] = clean_data[columns_to_clean].astype(float)

```

c) Change the categorical variables (Fuel_Type and Transmission) into numerical one hot
encoded value.

```python
import pandas as pd


# List of columns to one-time encode
categorical_columns = ['Fuel_Type', 'Transmission']

# Perform one-hot encoding using Pandas
for column in categorical_columns:
    dummies = pd.get_dummies(clean_data[column], prefix=column)
    clean_data = pd.concat([clean_data, dummies], axis=1)
    clean_data.drop(column, axis=1, inplace=True)

# Save the dataset with one-hot encoded variables
clean_data.to_csv('dataset_with_one_hot_encoding.csv', index=False)
```

```python
import pandas as pd
from datetime import datetime

# Loading  the dataset
clean_data = pd.read_csv('/content/cleaned_dataset.csv')

# Convert the 'Year' column to a datetime format.
clean_data['Year'] = pd.to_datetime(clean_data['Year'], format='%Y')

# Determine the vehicle's current age.
current_date = datetime.now()
clean_data['Current_Age'] = (current_date - clean_data['Year']).dt.days // 365

# Save the dataset with the new 'Current_Age' feature
clean_data.to_csv('dataset_with_current_age.csv', index=False)
```
